```python
# Multi-day Forecasting 
import pandas as pd

# ensure datetime index
prediction['date'] = pd.to_datetime(prediction.index)
prediction = prediction.set_index('date')

# make sure tavg is numeric
prediction['tavg'] = pd.to_numeric(prediction['tavg'], errors='coerce')

prediction = prediction.dropna()
```

```python
for lag in [1, 2, 3, 7]:
    prediction[f'tavg_lag_{lag}'] = prediction['tavg'].shift(lag)

prediction = prediction.dropna()
```

```python
X = prediction[['tavg_lag_1','tavg_lag_2','tavg_lag_3','tavg_lag_7']]
y = prediction['tavg']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, shuffle=False
)
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(
    n_estimators=150,
    random_state=42
)

model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error

preds = model.predict(X_test)
print("MAE:", mean_absolute_error(y_test, preds))
```

```python
last_known = prediction[['tavg_lag_1','tavg_lag_2','tavg_lag_3','tavg_lag_7']].iloc[-1:].copy()

future_preds = []

for day in range(7):
    next_temp = model.predict(last_known)[0]
    future_preds.append(next_temp)

    # update lags
    last_known.iloc[0, 3] = last_known.iloc[0, 2]
    last_known.iloc[0, 2] = last_known.iloc[0, 1]
    last_known.iloc[0, 1] = last_known.iloc[0, 0]
    last_known.iloc[0, 0] = next_temp
```

```python
prediction = prediction.reset_index(drop=True)

prediction['date'] = pd.date_range(
    start='2000-01-01',
    periods=len(prediction),
    freq='D'
)

prediction = prediction.set_index('date')

print(type(prediction.index))
print(prediction.index.min(), prediction.index.max())
```

```python
forecast_dates = pd.date_range(
    start=prediction.index[-1] + pd.Timedelta(days=1),
    periods=7
)

forecast = pd.DataFrame({
    'date': forecast_dates,
    'predicted_tavg': future_preds
})

print(forecast)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))
plt.plot(prediction.index[-100:], prediction['tavg'][-100:], label='Historical')
plt.plot(forecast['date'], forecast['predicted_tavg'], label='7-Day Forecast', marker='o')

plt.xlabel('Date')
plt.ylabel('Average Temperature')
plt.title('7-Day Temperature Forecast')
plt.legend()
plt.show()
```
